# DRY Design Principles
- DRY stands for Don't Repeat yoursef, and adhering to this principle can significantly improve code quality and reduce redundancy.
- Don't Repeat yourself (DRY) is a software development principle that encourages developers to avoid duplicating code in a system. The main idea behind DRY is to reduce redundancy and promote efficiency by ensuring that a particular piece of knowledge or logic exists in only on place within a codebase.
- When developoers adhere to the DRY principle, they aim to create reusable componenrs, functions, or modules that can be untilized in various parts of the codebase. This not only makes the code more maintainable but also minimizes the chances of errors since changes or updates only need to be made in one location.

## Key Features of the Dry principles
1. Code Reusability
    - DRY encourages developers to write code in a way that minimzes redundancy. Instead of duplicating code, developers should create reusable components, functions, or modules that can be shared and applied in multiple parts of the codebase.
2. Maintenance and Updates
    - By adhering to DRY, developers reduce the likelihood of errors and bugs that can arise from inconsistent updates. Since a particular piece of logic or knowledge exists in only one place, any changes or enhancements can be made in a centralized location, making maintenance more efficient.
3. Readability
    - DRY contributes to code readability by eliminating unnecessary repetition. When developers follow the principle, it becomes easier for others (or even themselves) to understand and navigate the codebase sicne there are fewer instances of similar or identical code scattered throughout.
4. Consistency
    - DRY promotoes consistency in the codebase. When a specific functionality is encapsulated in a single location, it ensures that all instances of that functionalitiy behave consistently. This is crucial for creating reliable and predictable software.
5. Reduced development time
6. Facilitates collaboration
7. Avoidance of Copy-Paste Errors
8. Testability

## How to implement the DRY principle
1. Use functions or Methods:
    - Encapsulate repetitive code into reusable functions or methods.

```python
class NumberSwapper:
    def swap(self, number: list[int]):
        temp = number[0]
        number[0] = number[1]
        number[1] = temp

if __name__ == "__main__":
    swapper = NumberSwapper()
    number = [1, 2]
    print(number)
    swapper.swap(number)
    print(number)
        
```
